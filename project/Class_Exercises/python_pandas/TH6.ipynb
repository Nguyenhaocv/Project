*Câu 7*

a) Thống kê số lượng `Digimon` của mỗi loại `Attribute`.

b) Digimon thuộc loại `Attribute` nào có điểm tấn công `Lv50 Atk` trung bình lớn nhất?

```python
digimons.Attribute.value_counts()
```

```python
dct = {}
for i in digimons.Attribute.unique():
  dct[i] = round(digimons['Lv50 Atk'][digimons.Attribute == i].mean(),2)

print(dct,'\n')
print('Attribute\'s Largest Average Lv 50 Atk is',max(dct, key=dct.get) + ':',max(dct.values()))
```

*Câu 8*

Thống kê trung bình của lượng máu (cột HP) của các Digimon theo từng `Type` và `Attribute`

```python
for i in digimons.Attribute.unique():
  for j in digimons.Type.unique():
    print(i,j,":",round(digimons['Lv 50 HP'][(digimons['Attribute'] == i) & (digimons['Type'] == j)].mean(),2))
```
